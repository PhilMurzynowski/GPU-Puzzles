## Puzzle 1: Map

Implement a "kernel" (GPU function) that adds 10 to each position of vector `a`
and stores it in vector `out`.  You have 1 thread per position.

**Warning** This code looks like Python but it is really CUDA! You cannot use
standard python tools like list comprehensions or ask for Numpy properties
like shape or size (if you need the size, it is given as an argument).
The puzzles only require doing simple operations, basically
+, *, simple array indexing, for loops, and if statements.
You are allowed to use local variables.
If you get an
error it is probably because you did something fancy :).

*Tip: Think of the function `call` as being run 1 time for each thread.
The only difference is that `cuda.threadIdx.x` changes each time.*

```python
def map_spec(a):
    return a + 10


def map_test(cuda):
    def call(out, a) -> None:
        local_i = cuda.threadIdx.x
        out[local_i] = a[local_i] + 10

    return call


SIZE = 4
out = np.zeros((SIZE,))
a = np.arange(SIZE)
problem = CudaProblem(
    "Map", map_test, [a], out, threadsperblock=Coord(SIZE, 1), spec=map_spec
)
problem.show()
```

```python
problem.check()
```

## Puzzle 8 - Shared

Implement a kernel that adds 10 to each position of `a` and stores it in `out`.
You have fewer threads per block than the size of `a`.

**Warning**: Each block can only have a *constant* amount of shared
 memory that threads in that block can read and write to. This needs
 to be a literal python constant not a variable. After writing to
 shared memory you need to call `cuda.syncthreads` to ensure that
 threads do not cross.

(This example does not really need shared memory or syncthreads, but it is a demo.)

```python
TPB = 4
def shared_test(cuda):
    def call(out, a, size) -> None:
        shared = cuda.shared.array(TPB, numba.float32)
        i = cuda.blockIdx.x * cuda.blockDim.x + cuda.threadIdx.x
        local_i = cuda.threadIdx.x

        if i < size:
            # not necessary to used shared, to build intuition
            shared[local_i] = a[i]
            cuda.syncthreads()
            out[i] = shared[local_i] + 10

    return call


SIZE = 8
out = np.zeros(SIZE)
a = np.ones(SIZE)
problem = CudaProblem(
    "Shared",
    shared_test,
    [a],
    out,
    [SIZE],
    threadsperblock=Coord(TPB, 1),
    blockspergrid=Coord(2, 1),
    spec=map_spec,
)
problem.show()
```

## Puzzle 9 - Pooling

Implement a kernel that sums together the last 3 position of `a` and stores it in `out`.
You have 1 thread per position. You only need 1 global read and 1 global write per thread.

*Tip: Remember to be careful about syncing.*

```python
def pool_spec(a):
    out = np.zeros(*a.shape)
    for i in range(a.shape[0]):
        out[i] = a[max(i - 2, 0) : i + 1].sum()
    return out


TPB = 8
def pool_test(cuda):
    def call(out, a, size) -> None:
        shared = cuda.shared.array(TPB, numba.float32)
        i = cuda.blockIdx.x * cuda.blockDim.x + cuda.threadIdx.x
        local_i = cuda.threadIdx.x # not needed

        # No shared, but used 3 global reads.

        # if i == 0: out[i] = a[i]
        # elif i == 1: out[i] = a[i-1] + a[i]
        # if i >= 2: out[i] = a[i-2] + a[i-1] + a[i]

        # Reading from block-shared memory is more efficient.
        # Could be 10s to 100s x faster. It's a cache.

        shared[i] = a[i]
        cuda.syncthreads()
        if i >= 2: out[i] = shared[i] + shared[i-1] + shared[i-2]
        elif i >= 1: out[i] = shared[i] + shared[i-1]
        else: out[i] = shared[i]


    return call


SIZE = 8
out = np.zeros(SIZE)
a = np.arange(SIZE)
problem = CudaProblem(
    "Pooling",
    pool_test,
    [a],
    out,
    [SIZE],
    threadsperblock=Coord(TPB, 1),
    blockspergrid=Coord(1, 1),
    spec=pool_spec,
)
problem.show()
```

## Puzzle 10 - Dot Product

Implement a kernel that computes the dot-product of `a` and `b` and stores it in `out`.
You have 1 thread per position. You only need 2 global reads and 1 global write per thread.

*Note: For this problem you don't need to worry about number of shared reads. We will
 handle that challenge later.*

```python
def dot_spec(a, b):
    return a @ b

TPB = 8
def dot_test(cuda):
    def call(out, a, b, size) -> None:
        shared = cuda.shared.array(TPB, numba.float32)

        i = cuda.blockIdx.x * cuda.blockDim.x + cuda.threadIdx.x
        local_i = cuda.threadIdx.x
        # FILL ME IN (roughly 9 lines)

        # Divide and conquer reduction.
        shared[i] = a[i] * b[i]
        cuda.syncthreads()
        # General reduction.
        for j in range(1, 3):
            if i % 2**j == 2**j-1: shared[i] = shared[i] + shared[i - 2**(j-1)]
            cuda.syncthreads()
        # Last reduction can write directly into the output.
        if i % 8 == 7: out[0] = shared[7] + shared[3]

        # Can also see from this reduction that it maybe better to have each thread
        # reduce many things at once? So we do not have idle threads
    return call


SIZE = 8
out = np.zeros(1)
a = np.arange(SIZE)
b = np.arange(SIZE)
problem = CudaProblem(
    "Dot",
    dot_test,
    [a, b],
    out,
    [SIZE],
    threadsperblock=Coord(SIZE, 1),
    blockspergrid=Coord(1, 1),
    spec=dot_spec,
)
problem.show()
```

## Puzzle 11 - 1D Convolution

Implement a kernel that computes a 1D convolution between `a` and `b` and stores it in `out`.
You need to handle the general case. You only need 2 global reads and 1 global write per thread.

```python
def conv_spec(a, b):
    out = np.zeros(*a.shape)
    len = b.shape[0]
    for i in range(a.shape[0]):
        out[i] = sum([a[i + j] * b[j] for j in range(len) if i + j < a.shape[0]])
    return out


MAX_CONV = 4
TPB = 8
TPB_MAX_CONV = TPB + MAX_CONV
def conv_test(cuda):
    def call(out, a, b, a_size, b_size) -> None:
        # allocating size TPB to handle general case.

        # These allocations are not for the general case.
        # shared_a = cuda.shared.array(TPB, numba.float32)
        # shared_b = cuda.shared.array(TPB, numba.float32)

        # These allocations are for the general case.
        shared_a = cuda.shared.array(TPB_MAX_CONV, numba.float32)
        shared_b = cuda.shared.array(MAX_CONV, numba.float32)


        i = cuda.blockIdx.x * cuda.blockDim.x + cuda.threadIdx.x
        local_i = cuda.threadIdx.x
        # convolution can be thought of as shifted and scaled summation so starting from there
        # - [update]: this is not a good way to think of it on GPU, too many writes
        # - [update]: was useful to diagram and write this from the perspective a single thread
        #             indexing into `out`.
        # - [update]: I think this is technically cross correlation not convolution.
        # why was this supposed to be 17 lines? much less here. Guess it is close.
        # - [update]: ah because this code is not yet truly general, should be looping.
        # - [update]: need to solve multi-block case. double test cases :)
        if i < a_size:

            # non-general
            # shared_a[i] = a[i]
            # if i < b_size: shared_b[i] = b[i]
            # cuda.syncthreads()
            # if i < a_size-2: out[i] = shared_b[0]*shared_a[i] + shared_b[1]*shared_a[i+1] + shared_b[2]*shared_a[i+2]
            # elif i < a_size-1: out[i] = shared_b[0]*shared_a[i] + shared_b[1]*shared_a[i+1]
            # else: out[i] = shared_b[0]*shared_a[i]

            # general in the case of one block only
            # shared_a[i] = a[i]
            # if i < b_size: shared_b[i] = b[i]
            # cuda.syncthreads()
            # acc = 0
            # for j in range(b_size): acc += shared_b[j]*shared_a[i+j] if i + j < a_size else 0
            # out[i] = acc

            # general in the case of multiple blocks for a.
            # We have a simplification that b will always be one block as we have set a MAX_CONV < TPB.
            # We can use TPB_MAX_CONV to perform the convolution across blocks by utilizing overlap.
            # The cleverness here is that in order to get two global reads per thread, we observe that
            # MAX_CONV at most half of the TPB size, and we can use this fact to get an additional MAX_CONV
            # reads of a from the threads that are not reading in b.

            # load in TPB block of A
            shared_a[local_i] = a[i]
            # use a bound to handle size edgecases
            bound = min(TPB_MAX_CONV, a_size)
            # load in B, and use rest of threads to load in more of A
            # use the smallest threads to read in b
            if local_i < b_size:
                shared_b[local_i] = b[local_i]
            # use the threads after b_size to read in more a
            elif i - b_size + TPB < bound:
                effective_i, effective_local_i = i - b_size, local_i - b_size
                shared_a[effective_local_i+TPB] = a[effective_i+TPB]
            cuda.syncthreads()
            # perform the convolution
            acc = 0
            for j in range(b_size): acc += shared_b[j]*shared_a[local_i+j] if local_i + j < bound else 0
            out[i] = acc

    return call


# Test 1

SIZE = 6
CONV = 3
out = np.zeros(SIZE)
a = np.arange(SIZE)
b = np.arange(CONV)
problem = CudaProblem(
    "1D Conv (Simple)",
    conv_test,
    [a, b],
    out,
    [SIZE, CONV],
    Coord(1, 1),
    Coord(TPB, 1),
    spec=conv_spec,
)
problem.show()
```

## Puzzle 12 - Prefix Sum

Implement a kernel that computes a sum over `a` and stores it in `out`.
If the size of `a` is greater than the block size, only store the sum of
each block.

We will do this using the [parallel prefix sum](https://en.wikipedia.org/wiki/Prefix_sum) algorithm in shared memory.
That is, each step of the algorithm should sum together half the remaining numbers.
Follow this diagram:

![](https://user-images.githubusercontent.com/35882/178757889-1c269623-93af-4a2e-a7e9-22cd55a42e38.png)

```python
TPB = 8
def sum_spec(a):
    out = np.zeros((a.shape[0] + TPB - 1) // TPB)
    for j, i in enumerate(range(0, a.shape[-1], TPB)):
        out[j] = a[i : i + TPB].sum()
    return out


def sum_test(cuda):
    def call(out, a, size: int) -> None:
        cache = cuda.shared.array(TPB, numba.float32)
        i = cuda.blockIdx.x * cuda.blockDim.x + cuda.threadIdx.x
        local_i = cuda.threadIdx.x
        # FILL ME IN (roughly 12 lines)

        if i < size:
            # Divide and conquer reduction.
            cache[local_i] = a[i]
            cuda.syncthreads()
            # General reduction.
            for j in range(1, 3):
                # Propogating this up (towards the 0-index) is actually critical for generalizability
                # as if it is variable length the only invariate is that we can reasonably assume it is length 1.
                if local_i % 2**j == 0: cache[local_i] = cache[local_i] + cache[local_i + 2**(j-1)]
                cuda.syncthreads()

        # Last reduction can write directly into the output.
        # But this only saves 1 operation out of logn so not that important.
        # Use only 1 thread, use the first thread to make it general for variable sizes.
        if local_i % 8 == 0: out[cuda.blockIdx.x] = cache[0] + cache[4]
    return call


# Test 1

SIZE = 8
out = np.zeros(1)
inp = np.arange(SIZE)
problem = CudaProblem(
    "Sum (Simple)",
    sum_test,
    [inp],
    out,
    [SIZE],
    Coord(1, 1),
    Coord(TPB, 1),
    spec=sum_spec,
)
problem.show()
```
